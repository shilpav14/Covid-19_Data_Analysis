```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import matplotlib.colors as mcolors
```

```python
df1 = pd.read_csv('AgeGroupDetails.csv')
df2 = pd.read_csv('HospitalBedsIndia.csv')
df3 = pd.read_csv('ICMRTestingLabs.csv')
df4 = pd.read_csv('IndividualDetails.csv')
df5 = pd.read_csv('StatewiseTestingDetails.csv')
```

```python
def plot_pie_chart(x,y,title):
    c = random.choices(list(mcolors.CSS4_COLORS.values()),k = 10)
    plt.figure(figsize = (20,15))
    plt.title(title, size = 20)
    plt.pie(y, colors = c)
    plt.legend(x, loc='best', fontsize = 15)
    plt.show()
```

```python
plot_pie_chart(df1['AgeGroup'], df1['TotalCases'], 'Age Distribution')
```

```python
def plot_bar_graph(x,y,title):
    plt.figure(figsize = (20, 12))
    plt.barh(x,y)
    plt.title(title, size = 20)
    plt.xticks(size = 15)
    plt.yticks(size = 20)
    plt.show()
    
#dropping 'All India' row value in 'State/UT'         
df2 = df2.drop(df2.index[36])
```

```python
plot_bar_graph(df2['State/UT'],df2['NumRuralHospitals_NHP18'], '# of Rural Hospitals in different regios')
```

```python
plot_bar_graph(df2['State/UT'], df2['NumRuralBeds_NHP18'], '# of Rural Beds in different regions')
```

```python
plot_bar_graph(df2['State/UT'], df2['NumUrbanHospitals_NHP18'], '# of Urban Hospitals in different regions')
```

```python
plot_bar_graph(df2['State/UT'], df2['NumUrbanBeds_NHP18'], '# of Urban Beds in different regions')
```

```python
plot_bar_graph(df2['State/UT'], df2['NumPublicBeds_HMIS'], '# of Public Beds in different regions')
```
